```python
import math
import matplotlib.pyplot as plt
class cos_graph:
  def generate_x_y(self):
    self.x=[]
    self.y=[]
    self.dx=(self.end-self.st)/(self.n-1)
    for i in range(self.n):
      x_=self.st+self.dx*i
      self.x.append(x_)
      self.y.append(math.cos(x_))
  def plot(self):
      plt.plot(self.x,self.y)
      plt.show()

g1=cos_graph()
g1.st=0
g1.end=2*math.pi
g1.n=10

g1.generate_x_y()
g1.plot()



class cos_graph:
  def setting(self,st,ed,n):
    self.st=st
    self.ed=ed
    self.n=n
  def generate_x_y(self):
    self.x=[]
    self.y=[]
    self.dx=(self.ed-self.st)/(self.n-1)
    for i in range(self.n):
      x_=self.st+self.dx*i
      self.x.append(x_)
      self.y.append(math.cos(x_))
  def plot(self):
      plt.plot(self.x,self.y)



plt.show()
g1=cos_graph()
g1.setting(0, 2*math.pi, 100)
g1.generate_x_y()
g1.plot()



g1=cos_graph()
g1.setting(0, 2*math.pi, 100)
g1.generate_x_y()
g1.plot()



g2=cos_graph()
g2.setting(0, 2*math.pi, 100)
g2.generate_x_y()
g2.plot()
```

```python
class text_static:
    def processLine(self, line, wordCounts):
        line = self.replacePunctuations(line)
        words = line.split()
        for word in words:
            if word in wordCounts:
                wordCounts[word] += 1
            else:
                wordCounts[word] = 1

    def replacePunctuations(self, line):
        for ch in line:
            if ch in ".,;:":
                line = line.replace(ch, " ")
        return line

    def wordCount(self, filename):
        f = open(filename, "r")
        lines = f.readlines()
        wordCounts = {}
        for line in lines:
            self.processLine(line.lower(), wordCounts)
        return wordCounts

    def arrange_pairs(self, wordCounts):
        pairs = list(wordCounts.items())
        items = [[x, y] for (y, x) in pairs]
        items.sort()
        items = {x: y for [y, x] in items}
        return items

T = text_static()
fn = "I_have_a_dream.txt"
wordCounts = T.wordCount(fn)
wordCounts = T.arrange_pairs(wordCounts)

print(wordCounts)
```
